```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics.pairwise import cosine_similarity
```

```python
subjects = ['Constitutional Law', 'Criminal Law', 'Contract Law', 'Torts', 'Family Law']
colleges = ['NLSIU', 'NLU Delhi', 'NALSAR', 'NLU Jodhpur', 'NLU Bhopal']
prep_levels = ['Beginner', 'Intermediate', 'Advanced']
styles = ['Visual', 'Reading', 'Interactive']


aspirants_data = {
    'aspirant_id': range(1, 51),
    'preferred_subjects': np.random.choice(subjects, 50),
    'target_college': np.random.choice(colleges, 50),
    'preparation_level': np.random.choice(prep_levels, 50),
    'learning_style': np.random.choice(styles, 50)
}
aspirants_df = pd.DataFrame(aspirants_data)


mentors_data = {
    'mentor_id': range(101, 111),
    'expertise_subject': np.random.choice(subjects, 10),
    'alumni_college': np.random.choice(colleges, 10),
    'mentoring_style': np.random.choice(styles, 10)
}
mentors_df = pd.DataFrame(mentors_data)
```

```python
aspirants_encoded = aspirants_df.copy()
mentors_encoded = mentors_df.copy()


aspirant_features = ['preferred_subjects', 'target_college', 'learning_style']
mentor_features = ['expertise_subject', 'alumni_college', 'mentoring_style']


for col in aspirant_features:
    le = LabelEncoder()
    aspirants_encoded[col] = le.fit_transform(aspirants_encoded[col])

for col in mentor_features:
    le = LabelEncoder()
    mentors_encoded[col] = le.fit_transform(mentors_encoded[col])
```

```python

aspirant_vectors = aspirants_encoded[aspirant_features].values
mentor_vectors = mentors_encoded[mentor_features].values
```

```python
similarity_matrix = cosine_similarity(aspirant_vectors, mentor_vectors)
```

```python
top_k = 3

recommendations = []

print("üîç Top 3 Mentor Recommendations:\n")
for i, similarities in enumerate(similarity_matrix):
    top_indices = np.argsort(similarities)[::-1][:top_k]
    aspirant_id = aspirants_df.loc[i, 'aspirant_id']
    print(f"Aspirant {aspirant_id} is best matched with mentors:")

    recs = []
    for idx in top_indices:
        mentor_id = mentors_df.loc[idx, 'mentor_id']
        score = similarities[idx]
        print(f"  - Mentor {mentor_id} (Similarity: {score:.2f})")
        recs.append((mentor_id, round(score, 2)))

    recommendations.append({'aspirant_id': aspirant_id, 'recommended_mentors': recs})
    print("-" * 60)
```
